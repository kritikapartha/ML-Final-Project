```python
from bs4 import BeautifulSoup
import json
from collections import defaultdict
```

```python
soup = BeautifulSoup(open("html.txt"))

soup = soup.find("div", {"class": "accordion accordion-with-arrow"})

#print(soup.prettify())

headers_dict = {}
ids = defaultdict(list)

#get headers
headers = soup.children
for header in headers:

  #get header names
  header_name = header.find("a", {"class": "main-parents"}).get_text()
  #print(header_name)

  header_id = header.find("a")['data-uri']
  ids[header_name].append(header_id)

  sub_headers_dict = {}

  #get subheaders
  sub_headers = header.find("ul", {"class": "rlist accordion-tabbed__content classification-system__ccs-flat-list child-list"})
  for sub_header in sub_headers:

    #get subheader name
    sub_header_name = sub_header.find("a", {"class": "list-second"}).get_text()
    #print("   "  + sub_header_name)

    sub_header_id = sub_header.find("a")['data-uri']
    ids[sub_header_name].append(sub_header_id)

    children_one_dict = {}

    #get first children
    children_one = sub_header.find("ul", {"class": "rlist child-list"})
    for child_one in children_one:

      #get first children name
      child_one_name = child_one.find("a").get_text()
      #print("      " + child_one_name)

      child_one_id = child_one.find("a")['data-uri']
      ids[child_one_name].append(child_one_id)

      children_two_dict = {}

      #get second children
      children_two = child_one.find("ul", {"class": "rlist child-list"})
      if children_two is not None:
        for child_two in children_two:

          #get second children name
          child_two_name = child_two.find("a").get_text()
          #print("         " + child_two_name)

          child_two_id = child_two.find("a")['data-uri']
          ids[child_two_name].append(child_two_id)

          children_three_dict = {}

          #get third children
          children_three = child_two.find("ul", {"class": "rlist child-list"})
          if children_three is not None:
            for child_three in children_three:

              #get third children name
              child_three_name = child_three.find("a").get_text()
              #print("            " + child_three_name)

              child_three_id = child_three.find("a")['data-uri']
              ids[child_three_name].append(child_three_id)

              children_four_dict = {}

              #get fourth children name
              children_four = child_three.find("ul", {"class": "rlist child-list"})
              if children_four is not None:
                for child_four in children_four:

                  #get fourth children name
                  child_four_name = child_four.find("a").get_text()
                  #print("               " + child_four_name)

                  child_four_id = child_four.find("a")['data-uri']
                  ids[child_four_name].append(child_four_id)

                  children_four_dict[child_four_name] = None

              children_three_dict[child_three_name] = children_four_dict

          children_two_dict[child_two_name] = children_three_dict

      children_one_dict[child_one_name] = children_two_dict

    sub_headers_dict[sub_header_name] = children_one_dict

  headers_dict[header_name] = sub_headers_dict

id_json = json.dumps(ids, indent = 4)
print(id_json)

index_json = json.dumps(headers_dict, indent = 4)
#print(pretty_json)
```
